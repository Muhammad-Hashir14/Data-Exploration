```python
def calculate(s1 , s2 ,s3, s4 , s5):
    addit = s1 + s2 + s3 + s4 + s5
    per = addit/500 * 100
    grade = ''
    if per > 80 and per <= 100:
        grade = 'A+'
    elif per > 70 and per <= 80:
        grade = 'A'
    elif per > 60 and per <= 70:
        grade = 'B'
    elif per > 50 and per <=60:
        grade = 'C'
    elif per > 40 and per <= 50:
        grade = 'D'
    else:
        grade = 'F'
    return addit , per , grade
```

```python
df[['Obtained Marks','Percentage','Grade']] = df[['math score','reading score','writing score','Programming score','Theory score']].apply(lambda x:calculate(x[0],x[1],x[2],x[3],x[4]) ,axis=1,result_type='expand')
```
